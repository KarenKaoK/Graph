# Short path : 六度理論


```python
import gzip
import random
import networkx as nx
import matplotlib.pyplot as plt 

# Convert the data to a NetworkX graph
graph = nx.read_edgelist('./data/facebook_combined.txt', delimiter=' ', nodetype=int)

print(nx.info(graph))

sp = nx.spring_layout(graph)

plt.figure(figsize=(30, 20))
nx.draw_networkx(graph, pos=sp, with_labels=False, node_size=5, node_color='r')
plt.savefig('graph.png', dpi=800)

```
