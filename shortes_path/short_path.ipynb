# Short path : 六度理論


```python
import gzip
import random
import networkx as nx
import matplotlib.pyplot as plt 

# Convert the data to a NetworkX graph
graph = nx.read_edgelist('./data/facebook_combined.txt', delimiter=' ', nodetype=int)

print(nx.info(graph))

sp = nx.spring_layout(graph)

plt.figure(figsize=(30, 20))
nx.draw_networkx(graph, pos=sp, with_labels=False, node_size=5, node_color='r')
plt.savefig('graph.png', dpi=800)

```

```python
# 隨機抽取一個節點
node = random.choice(list(graph.nodes()))

# 獲取該節點的鄰居
neighbors = list(nx.neighbors(graph, node))

# 創建一個新的圖，只包含該節點及其鄰居
subgraph = graph.subgraph([node] + neighbors)

# 可視化子圖
sp = nx.spring_layout(subgraph)
plt.figure(figsize=(10, 10))
nx.draw_networkx(subgraph, pos=sp, with_labels=False, node_size=5, node_color='r')
plt.savefig('subgraph.png', dpi=300)
```

# 計算圖形的基本屬性


```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.read_edgelist('./data/facebook_combined.txt', delimiter=' ', nodetype=int)


node_count = graph.number_of_nodes()
edge_count = graph.number_of_edges()
average_degree = sum(dict(graph.degree()).values()) / node_count
density = nx.density(graph)

print(f"節點數: {node_count}")
print(f"邊數: {edge_count}")
print(f"平均度數: {average_degree:.2f}")
print(f"網路密度: {density:.4f}")
```

# 六度分隔

```python
import networkx as nx

# select one node 
source_node = 0  

# Dijkstra 
lengths = nx.single_source_dijkstra_path_length(graph, source_node)


for target_node, length in lengths.items():
    print(f"從節點{source_node}到節點{target_node}的最短路徑長是：{length}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt


lengths = dict(nx.shortest_path_length(graph))


path_lengths = []
for source, targets in lengths.items():
    path_lengths.extend(targets.values())


length_distribution = {}
for length in path_lengths:
    if length in length_distribution:
        length_distribution[length] += 1
    else:
        length_distribution[length] = 1


total_length = sum(length * count for length, count in length_distribution.items())
total_count = sum(length_distribution.values())
average_path_length = total_length / total_count


plt.figure(figsize=(10, 5))
plt.bar(length_distribution.keys(), length_distribution.values(), color='b')
plt.title('Path Length Distribution in Ego-Facebook Network')
plt.xlabel('Path Length')
plt.ylabel('Frequency')
plt.xticks(range(max(length_distribution.keys()) + 1))
plt.grid(axis='y', linestyle='--')

plt.axvline(x=average_path_length, color='r', linestyle='--', label=f'Average Path Length: {average_path_length:.2f}')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import random


nodes = list(graph.nodes())
source = random.choice(nodes)
target = random.choice(nodes)
while target == source:  # 確保終點和起點不同
    target = random.choice(nodes)

# 計算最短路徑
path = nx.shortest_path(graph, source=source, target=target)

# 可視化最短路徑
pos = nx.spring_layout(graph)  # 為圖中的節點生成一個佈局
# 繪製整個網絡，淡化顯示
nx.draw_networkx_nodes(graph, pos, node_size=10, alpha=0.3)
nx.draw_networkx_edges(graph, pos, alpha=0.1)
# 繪製最短路徑
nx.draw_networkx_nodes(graph, pos, nodelist=path, node_color='r', node_size=50)
nx.draw_networkx_edges(graph, pos, edgelist=[(path[i], path[i+1]) for i in range(len(path)-1)], width=2, edge_color='r')

plt.title(f"Shortest path from {source} to {target}")
plt.axis('off')  # 不顯示軸
plt.show()
```

```python
from pyvis.network import Network
import networkx as nx
import random


nodes = list(graph.nodes())
source = random.choice(nodes)
target = random.choice(nodes)
while target == source:
    target = random.choice(nodes)


path = nx.shortest_path(graph, source=source, target=target)


net = Network()


net.from_nx(graph)


for node in path:
    net.get_node(node)["color"] = "red" if node in [source, target] else "orange"
    net.get_node(node)["size"] = 15

for i in range(len(path)-1):
    edge = net.get_edges(path[i], path[i+1])
    edge["color"] = "red"
    edge["width"] = 5

net.show('example.html')
```
