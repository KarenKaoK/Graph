# degree centrality

```python
# 使用 範例資料 twitter

import pandas as pd 
import networkx as nx 
import matplotlib.pyplot as plt 

nodes = pd.read_csv('../dataset/twitter/social-nodes.csv')
relationships = pd.read_csv('../dataset/twitter/social-relationships.csv')

G = nx.Graph()

for idx, row in nodes.iterrows():
    G.add_node(row['id'], attr_dict=row.to_dict())

for idx, row in relationships.iterrows():
    G.add_edge(row['src'], row['dst'], weight= row.get('weight', 1))

plt.figure(figsize=(10,8))
nx.draw(G, with_labels=True, node_size=700, node_color='skyblue', font_size=10)
plt.show()
```
